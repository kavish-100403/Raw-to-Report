```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#First we will read the file
df=pd.read_csv("global_temp.csv")
```

```python
#On checking, we can see the spacings in column names so we will remove it first
df.columns = df.columns.str.strip()
df.columns
```

```python
#We will now check the number of entrees in each column
print(df.count())

#now, we will check if there are any duplicates, if any it will remove them
df=df.drop_duplicates()
print(df.count())
```

```python
# df['Date'] = (df['Date'].astype(str).str.replace(r'\s+', ' ', regex=True).str.strip())

#Apr-50
df['Date']=df['Date'].str.replace(r'([A-Za-z]{3})-(\d{2})$' ,r'\1-19\2', regex= True)
#Apr-50
df['Date']=df['Date'].str.replace(r'([A-Za-z]{3}) (\d{2})$' ,r'\1-19\2', regex= True)
#25-Apr
df['Date']=df['Date'].str.replace(r'(\d{2})-([A-Za-z]{3})$' ,r'\2-19\1', regex= True)
#25 Apr
df['Date']=df['Date'].str.replace(r'(\d{2}) ([A-Za-z]{3})$' ,r'\2-19\1', regex= True)


df['Date'] = (df['Date'].astype(str).str.strip())

df['Date'] = pd.to_datetime(df['Date'],errors='coerce',format='mixed',dayfirst=False)

df['Date'] = df['Date'].dt.strftime('%Y-%m-%d')
print(df['Date'].dtype)
```

```python
type(df['Date'].iloc[1698])
df['Date'].iloc[1698]
```

```python
#Handling the Anomalies of the 
df['Temperature_Anomaly']=df['Temperature_Anomaly'].astype(str).str.strip()
df['Temperature_Anomaly']=df['Temperature_Anomaly'].str.replace(r"[,]", ".", regex=True)
df['Temperature_Anomaly']=df['Temperature_Anomaly'].str.replace(r"[^\d.-]", "", regex=True)
df['Temperature_Anomaly'].iloc[10]

#Now, we will convert it into float as they are in string format
print(type(df['Temperature_Anomaly'].iloc[1]))

#For converting to float we will use pandas function and we are using erros='coerce' as there are missing values present 
df['Temperature_Anomaly']= pd.to_numeric(df['Temperature_Anomaly'], errors='coerce')
print(df['Temperature_Anomaly'].dtype)

# print(type(df['Temperature_Anomaly'].iloc[1]))

```

```python
#checking the progress
filename='1.csv'
df.to_csv(filename, index=False)
```
